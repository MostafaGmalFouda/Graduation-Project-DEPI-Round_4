```python
# Imports 
from Phase_1.DataLoader import DataLoader
from Phase_1.DataValidator import DataValidator
from Phase_1.DataPreprocessor import DataPreprocessor
from Phase_1.OutlierHandler import OutlierHandler
from Phase_1.ReportGenerator import ReportGenerator
from Phase_1.EDAPipeline import EDAPipeline
```

```python
# check DataLoader
file_path = "heart.csv"

data = DataLoader(file_path)
data.load()
data = data.get_data()
data
```

```python
# check DataValidator
validator = DataValidator(data)
validator.check_nulls()
validator.check_types()
validator.check_duplicates()
validator.report_issues()
```

```python
# check Datapreprocessor
preprocessor = DataPreprocessor(data)
preprocessor.handle_nulls()
preprocessor.convert_types()
preprocessor.remove_duplicates()
preprocessor.handle_outliers()
preprocessor.get_clean_data()
```

```python
# check OutlierHandler
outlier_handler = OutlierHandler(data)
outlier_handler.detect_iqr()
outlier_handler.detect_zscore()
outlier_handler.remove_outliers()
outlier_handler.cap_outliers()
```

```python
# check ReportGenerator
report_generator = ReportGenerator(data)
report_generator.summary_stats()
report_generator.correlation_matrix()
report_generator.missing_values()
report_generator.manual_report()
report_generator.auto_report()
report_generator.generate_report()
```

```python
# Check EDAPipeline
pipeline = EDAPipeline("heart.csv")
pipeline.run_pipeline()
```
