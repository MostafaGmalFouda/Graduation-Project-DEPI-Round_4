```python
from Phase_1.DataLoader import DataLoader
from Phase_1.DataPreprocessor import DataPreprocessor
from Phase_1.OutlierHandler import OutlierHandler
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

# Load your data
loader = DataLoader("heart.csv")
loader.load()
data = loader.get_data()

# Preprocess
preprocessor = DataPreprocessor(data)
preprocessor.handle_nulls()
preprocessor.convert_types()
preprocessor.remove_duplicates()
clean_data = preprocessor.get_clean_data()

# Handle outliers
handler = OutlierHandler(clean_data)
clean_data = handler.cap_outliers('iqr')

print("✅ Phase 1 done — clean_data ready")
print(clean_data.shape)
```

```python
from Phase_2.DataVisualizer import DataVisualizer

# Option A: Full auto-detect (recommended first try)
viz = DataVisualizer(clean_data)

# Option B: With schema override (if auto-detect got something wrong)
# viz = DataVisualizer(clean_data, schema={"year": "cat", "income": "num"})
```

```python
# Summary dashboard — shows ALL columns at once
viz.generate_summary_dashboard()

# Missing values matrix (skips automatically if no nulls)
viz.plot_missing_values_matrix()

# Correlation heatmap (numeric columns only)
viz.plot_correlation_heatmap()
```

```python
# Replace 'col_A' and 'col_B' with your actual numeric column names
num_cols = viz._get_cols_by_type("num")
print("Numeric columns available:", num_cols)

# 2D Scatter
viz.plot_scatter_2d(num_cols[0], num_cols[1])

# 2D Scatter colored by a category
cat_cols = viz._get_cols_by_type("cat")
viz.plot_scatter_2d(num_cols[0], num_cols[1], color_col=cat_cols[0])

# 3D Scatter (need at least 3 numeric columns)
if len(num_cols) >= 3:
    viz.plot_scatter_3d(num_cols[0], num_cols[1], num_cols[2])

# Joint plot — try kind='scatter', 'kde', 'hex', 'reg'
viz.plot_joint_plot(num_cols[0], num_cols[1], kind="kde")
```

```python
cat_cols = viz._get_cols_by_type("cat")
num_cols = viz._get_cols_by_type("num")
print("Categorical columns available:", cat_cols)

# Stacked bar
viz.plot_stacked_bar(cat_cols[0], cat_cols[1])

# Stacked bar normalized (proportions)
viz.plot_stacked_bar(cat_cols[0], cat_cols[1], normalize=True)

# Cross-tabulation heatmap
viz.plot_cross_tabulation(cat_cols[0], cat_cols[1])

# Violin plot — numeric distribution split by category
viz.plot_violin_plot_by_category(num_cols[0], cat_cols[0])

# Facet grid — multiple numeric cols faceted by a category
viz.plot_facet_grid(num_cols[:2], cat_cols[0])

# Bubble chart — x, y, size, color
viz.plot_bubble_chart(x=num_cols[0], y=num_cols[1], size=num_cols[2], color=cat_cols[0])
```

```python
# Check what schema was detected
print("Schema:", viz.schema)

# Check column names
print("Columns:", clean_data.columns.tolist())
print("Dtypes:\n", clean_data.dtypes)

# Check plots folder
import os
print("Saved plots:", os.listdir(viz.plots_dir))
```
